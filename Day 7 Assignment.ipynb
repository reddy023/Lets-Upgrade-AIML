## Outlier removal

```python
Q1 = attrition.quantile(0.25)
Q3 = attrition.quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
```

```python
attrition = attrition[~((attrition <lower) |(attrition > upper)).any(axis = 1)]
```
